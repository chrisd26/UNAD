```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt


#Imprime primeros 5 datos
df = pd.read_excel('Adidas US Sales Datasets.xlsx', sheet_name='Data Sales Adidas')
print(df.head())
```

```python
#Definimos los datos en un array
UnitsSold = df['Units Sold'].values.reshape(-1,1) 
PriceperUnit = df['Price per Unit'].values.reshape(-1,1)
UnitsSold
PriceperUnit
```

```python
linear_regressor = LinearRegression()
regresion = linear_regressor.fit(UnitsSold, PriceperUnit ) #creamos la regresion lineal
PriceperUnit_pred = linear_regressor.predict(UnitsSold) #predice el valor promedio

regresion
PriceperUnit_pred
```

```python
r2 = r2_score(PriceperUnit, PriceperUnit_pred) #Se calcula el valor de r2
r_2 = '%.4f' % r2 
label_reg = r'$r^2$ = ' + '' + str(r_2)
print('coeficiente r2 de la regresion: ' + '' + label_reg)
```

```python
fig = plt.figure()
plt.grid()
plt.scatter(df['Units Sold'], df['Price per Unit'], label=' 'r'$\nabla^2 \rho$ promedio')
plt.plot(UnitsSold, PriceperUnit_pred, color='royalblue') 
# texto = plt.text(25, 9, label_reg, color='royalblue', fontsize = 18)
# plt.scatter(df['Units Sold'], df['Price per Unit'], color 'darkorange', label =' 'r'$\nabla^2 \rho$ PriceperUnit')
# plt.plot(UnitsSold, PriceperUnit_pred, color = 'darkorange')


```
